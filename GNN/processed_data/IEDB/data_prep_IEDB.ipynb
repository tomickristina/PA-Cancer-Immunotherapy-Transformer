```python
import pandas as pd
# Load the CSV file
# GNN/processed_data/IEDB/MHCI_IEDB_beta_export.csv

file_path = 'MHCI_IEDB_beta_export.csv'
df = pd.read_csv(file_path, low_memory = False)

# # Inspect the first few rows and column names
# print("First few rows of the CSV:")
# print(df.head())
# print("\nColumn names in the CSV:")
# print(df.columns)

# Delete columns (replace 'column_to_delete1', 'column_to_delete2' with actual column names)
columns_to_delete = ['Assay - MHC Allele Names', 'Chain 1 - Type',
       'Chain 1 - Curated V Gene', 'Chain 1 - Curated J Gene',
       'Chain 1 - Protein Sequence', 'Chain 1 - CDR3 Curated',
       'Chain 1 - CDR1 Curated', 'Chain 1 - CDR2 Curated', 'Chain 2 - Type',
       'Chain 2 - Curated V Gene', 'Chain 2 - Curated J Gene',
       'Chain 2 - Protein Sequence',
       'Chain 2 - CDR1 Curated', 'Chain 2 - CDR2 Curated']
df = df.drop(columns=columns_to_delete)

# Rename columns (replace 'old_column_name': 'new_column_name' with actual names)
columns_to_rename = {'Epitope - Name': 'Epitope', 'Chain 2 - CDR3 Curated': 'CDR3.beta'}
df = df.rename(columns=columns_to_rename)
df_cleaned = df.dropna()
# Save the modified DataFrame to a new CSV file
output_path = 'IEDB.csv'
df_cleaned.to_csv(output_path, index=False)

# print(f"\nModified CSV saved as {output_path}")
```
